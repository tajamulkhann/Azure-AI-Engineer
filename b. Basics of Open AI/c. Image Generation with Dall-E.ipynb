## Dall-E Image Generation Model

Generating Images from Prompts using Dall-E.

`client.images.generate(model="dall-e-3", prompt="a Black furry Dog", size="1024x1024", quality="standard", n=1)`

```python
import openai
from openai import OpenAI
import os
from dotenv import load_dotenv
load_dotenv()

# For Environments
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")

# Get API key
api_key = os.getenv("OPENAI_API_KEY")

# Initialize OpenAI client
client = OpenAI()

response = client.images.generate(
  model="dall-e-3",
  prompt="a Black furry Dog with black eyes with a dog collar and a white cat with blue eyes with a necklace",
  size="1024x1024",
  quality="standard",
  n=1,
)

image_url = response.data[0].url
print(image_url)
```
